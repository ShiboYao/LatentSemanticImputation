```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.rc('pdf', fonttype=42)
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
f = plt.figure(figsize=(4, 3))

a = pd.read_csv('heat.csv', index_col=0)
plt.xlabel("Minimum Word Frequency")
plt.ylabel("k")
ax = plt.gca()
ax.set_yticks(range(len(a.index)))
ax.set_xticks(range(0, len(a.columns), 10))
ax.set_yticklabels(a.index.values)
ax.set_xticklabels(range(len(a.columns), 0, -10))
plt.imshow(a, cmap='hot', interpolation='nearest', aspect='auto')

clb = plt.colorbar()
clb.ax.set_title('Accuracy')
plt.tight_layout()
plt.show()
f.savefig("heat.pdf")
```

```python
kset = [3,5,8,10,15,20,30,50]
a = []
for k in kset:
    a.append(pd.read_csv('result_PQ'+str(k)+'_freq.csv', index_col=0).iloc[:,-40:])

f, axn = plt.subplots(4, 2, sharex=True, sharey=True, figsize=(7,11.5)) #16 7
plt.xlabel("x")

cbar_ax = f.add_axes([.9,0.35,0.015,0.3])


for i, ax in enumerate(axn.flat):
    sns.heatmap(a[i], cmap='plasma', ax=ax,cbar=i == 0,vmin=0.13, vmax=.58, cbar_ax=None if i else cbar_ax,
               linewidths=0.0, rasterized=True)
    ax.set_title('$\delta=$%d'%kset[i])
    ax.set_yticks(range(len(a[i].index)))
    ax.set_xticks(range(0, len(a[i].columns), 10))
    ax.set_yticklabels(a[i].index.values)
    ax.set_xticklabels(range(len(a[i].columns), 0, -10))
    
ax.set_ylabel('k')    
ax.set_xlabel('Minimum Word Frequency')
f.tight_layout(rect=[0, 0, .9, 1])
```

```python
f.savefig("deltaheat.pdf")
```

```python
b = pd.read_csv('ksens.csv', index_col=0)
f = plt.figure(figsize=(4, 3))
index = b.index.tolist()
y = b.columns.values
for i in range(len(b)):
    plt.plot(b.iloc[i], label = '$\delta$=%d' %index[i])
    plt.scatter(y, b.iloc[i], s=20, marker='x', label=None)
plt.xlabel("k")
plt.ylabel("kNN Accuracy")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
f.savefig("ksens.pdf")
```

```python
c = pd.read_csv('etasens.csv', index_col=0)
f = plt.figure(figsize=(4, 3))
index = c.index.tolist()
for i in range(len(c)):
    plt.plot(c.iloc[i], label = '$\eta$=1e%d' %np.log10(index[i]))
    plt.scatter(y, c.iloc[i], s=20, marker='x', label=None)
plt.xlabel("k")
plt.ylabel("kNN Accuracy")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
f.savefig("etasens.pdf")
```

```python
w = pd.read_csv('W.csv', index_col=0)
arr0 = w.values.copy()
arr1 = w.values.copy()
I = np.eye(arr1.shape[0])
for i in range(100):
    arr1[i] = I[i]
```

```python
e0, v0 = np.linalg.eig(arr0)
e1, v1 = np.linalg.eig(arr1)
```

```python
fig = plt.figure(figsize=(3,2))
plt.scatter(range(len(e0)), sorted(abs(e0), reverse=True), s=0.3)
plt.yticks(np.linspace(0,1, 6))
plt.ylabel('$|\lambda|$')
plt.tight_layout()
plt.show()
fig.savefig('eigw.pdf')
```

```python
fig = plt.figure(figsize=(3,2))
plt.scatter(range(len(e1)), sorted(abs(e1), reverse=True), s=0.3)
plt.ylabel('$|\lambda\'|$')
plt.yticks(np.linspace(0,1, 6))
plt.tight_layout()
plt.show()
fig.savefig('eigw1.pdf')
```
