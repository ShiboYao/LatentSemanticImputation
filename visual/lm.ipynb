```python
import pandas as pd
import matplotlib
matplotlib.rc('pdf', fonttype=42)
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("split_train.txt", sep=' ', index_col=0, header=None)
valid = pd.read_csv("split_valid.txt", sep=' ', index_col=0, header=None)
name = train.index.tolist()
name = [n.replace('_', '+') for n in name]
train.index = name
valid.index = name
sel = ['google', 'google+aff', 'fast', 'fast+aff', 'glove', 'glove+aff', 'self', 'self+google', 'self+fast', 'self+glove', 'self+aff']
train = train.loc[sel]
valid = valid.loc[sel]
```

```python
def PLOT(ax, ind):
    plt.scatter(train.columns, train.iloc[ind], marker='^')
    plt.scatter(valid.columns, valid.iloc[ind], marker='x')
    plt.title(train.index[ind])
    ax.tick_params(labelleft=False)
    ax.set_axisbelow(True)
    ax.set_yticks(range(11,15), minor=True)
    ax.yaxis.grid(which='both', color='lightgrey', linestyle='dashed')
```

```python
fig = plt.figure(figsize=(16,6))
grid = plt.GridSpec(2, 30, wspace=0.3, hspace=0.4)

ax1 = plt.subplot(grid[0, 0:5])
plt.yticks(range(0,26, 5))
plt.scatter(train.columns, train.iloc[0], marker='^')
plt.scatter(valid.columns, valid.iloc[0], marker='x')
plt.ylabel('PP')
plt.xlabel('epoch')
plt.title(train.index[0])
ax1.set_axisbelow(True)
ax1.set_yticks(range(11,15), minor=True)
ax1.yaxis.grid(which='both', color='lightgrey', linestyle='dashed')

ax2 = plt.subplot(grid[0, 5:10], sharey = ax1)
PLOT(ax2, 1)

ax3 = plt.subplot(grid[0, 10:15], sharey = ax1)
PLOT(ax3, 2)

ax4 = plt.subplot(grid[0, 15:20], sharey = ax1)
PLOT(ax4, 3)

ax5 = plt.subplot(grid[0, 20:25], sharey = ax1)
PLOT(ax5, 4)

ax6 = plt.subplot(grid[0, 25:30], sharey = ax1)
PLOT(ax6, 5)


ax7 = plt.subplot(grid[1, 2:7], sharey = ax1)
plt.scatter(train.columns, train.iloc[6], marker='^')
plt.scatter(valid.columns, valid.iloc[6], marker='x')
plt.ylabel('PP')
plt.xlabel('epoch')
plt.title(train.index[6])
ax7.set_axisbelow(True)
ax7.set_yticks(range(11,15), minor=True)
ax7.yaxis.grid(which='both', color='lightgrey', linestyle='dashed')

ax8 = plt.subplot(grid[1, 7:12], sharey = ax1)
PLOT(ax8, 7)

ax9 = plt.subplot(grid[1, 12:17], sharey = ax1)
PLOT(ax9, 8)

ax10 = plt.subplot(grid[1, 17:22], sharey = ax1)
PLOT(ax10, 9)

ax11 = plt.subplot(grid[1, 22:27], sharey = ax1)
PLOT(ax11, 10)

#fig.legend( ax1, ['train', 'validation'], loc = (0.5, 0), ncol=5 )
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, ['train', 'validation'] , loc = (0.83, 0.2), ncol=1 )

plt.show()
```
